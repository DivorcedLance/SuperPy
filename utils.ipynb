```python
class Cajero:
    def __init__(self, id, nombre, entrada, salida):
        self.id : int = id
        self.nombre : str = nombre
        self.entrada = entrada
        self.salida = salida
        self.caja = None

    def __str__(self):
        return f"{self.nombre} - {self.entrada} - {self.salida} - {self.caja}"
    
class Caja:
    def __init__(self, id, nombre):
        self.nro : int = id
        self.cajeros : list[Cajero] = []

    def __str__(self):
        return f"{self.nro} - {self.cajeros}"
    
    def agregar_cajero(self, cajero):
        self.cajeros.append(cajero)
        cajero.caja = self
```

```python
import pandas as pd
data_path = "schedule.csv"
df = pd.read_csv(data_path, sep=";")
```

```python
day_schedule_df = df[["Nombres", "Sabado"]]
# where sea diferente de DIA DE DESCANSO o VACACIONES
day_schedule_df = day_schedule_df[day_schedule_df["Sabado"] != "DIA DE DESCANSO"]
day_schedule_df = day_schedule_df[day_schedule_df["Sabado"] != "VACACIONES"]
```

```python
def threat_cell(cell):
    if cell in ["DIA DE DESCANSO", "VACACIONES"]:
        return cell, cell
    else:
        entrada = cell.split(" - ")[0]
        salida = cell.split(" - ")[1]
        return entrada, salida

# use threat_cell function to threat the cell

day_schedule_df["Entrada"], day_schedule_df["Salida"] = zip(*day_schedule_df["Sabado"].map(threat_cell))

# drop lunes
day_schedule_df = day_schedule_df.drop(columns=["Sabado"])

day_schedule_df["Entrada"] = pd.to_datetime(day_schedule_df["Entrada"])
day_schedule_df["Salida"] = pd.to_datetime(day_schedule_df["Salida"])
```

```python
def asignar_cajas(nrCajas, day_schedule_df):
    caja_matrix = [[] for _ in range(0, nrCajas+1)]
    cajas = [None for i in range(0, nrCajas+1)]
    asignados = []

    for i, (id, row) in enumerate(day_schedule_df.iterrows()):
        if (i == 0):
            cajas[1] = row
            asignados.append((row["Entrada"], row["Salida"], 1, id))
            caja_matrix[1].append(f"{row['Nombres']} - {row['Entrada'].strftime('%I:%M%p')} - {row['Salida'].strftime('%I:%M%p')}")
            continue
        for j, asignado in enumerate(asignados):
            if (asignado is None):
                continue
            if (row["Entrada"] >= asignado[1]):
                cajas[asignado[2]] = row
                asignados.append((row["Entrada"], row["Salida"], asignado[2], id))
                caja_matrix[asignado[2]].append(f"{row['Nombres']} - {row['Entrada'].strftime('%I:%M%p')} - {row['Salida'].strftime('%I:%M%p')}")
                asignados.pop(j)
                break
        else:
            for j, caja in enumerate(cajas):
                if (caja is None):
                    cajas[j] = row
                    asignados.append((row["Entrada"], row["Salida"], j, id))
                    caja_matrix[j].append(f"{row['Nombres']} - {row['Entrada'].strftime('%I:%M%p')} - {row['Salida'].strftime('%I:%M%p')}")
                    break
            else:
                print(f"No hay cajas disponibles, {row}")
    return caja_matrix
```

```python
def matriz_to_excel(matriz, nombre_archivo):
    # Crear un DataFrame a partir de la matriz
    df = pd.DataFrame(matriz)
    
    # Guardar el DataFrame en un archivo Excel
    df.to_excel(nombre_archivo, index=False, header=False)
```

```python
def calcular_tiempos_muertos(caja_matrix):
    tiempos_muertos = pd.to_datetime("00:00AM")
    for caja in caja_matrix:
        for i in range(1, len(caja)):
            entrada = caja[i-1].split(" - ")[2]
            salida = caja[i].split(" - ")[1]
            tiempo_muerto = pd.to_datetime(salida) - pd.to_datetime(entrada)
            tiempos_muertos += tiempo_muerto
    return tiempos_muertos
```

```python
nrMax = 22
cajasRange = (5, nrMax)
minTime = pd.to_datetime("00:00AM")
minCajas = 0
minMatrix = []

for nr_cajas in range(*cajasRange):
    print(f"Probando con {nr_cajas} cajas")
    caja_matrix = asignar_cajas(nr_cajas, day_schedule_df)
    tiempo_muerto = calcular_tiempos_muertos(caja_matrix)
    if (nr_cajas == cajasRange[0]):
        minTime = tiempo_muerto
        minCajas = nr_cajas
        minMatrix = caja_matrix
    else:
        if (tiempo_muerto < minTime):
            minTime = tiempo_muerto
            minCajas = nr_cajas
            minMatrix = caja_matrix

matriz_to_excel(minMatrix, "cajas.xlsx")
```
